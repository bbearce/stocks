```python
import os
import pdb
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
cb = pd.read_csv("Individual-Positions-2024-06-20-160845.csv")
cb
```

```python

# Define the stock symbols and the time period
stocks = cb.symbol.to_list()
start_date = '2015-01-01'
# start_date = '2016-01-01'
# start_date = '2017-01-01'
# start_date = '2017-01-01'
# start_date = '2019-01-01'
end_date = f"{datetime.date.today().year}-{datetime.date.today().month}-{datetime.date.today().day}"
```

```python

# Function to fetch stock data
def fetch_stock_data(stock, start, end):
    return yf.download(stock, start=start, end=end)['Close']

# Fetch data for all stocks
data = {stock: fetch_stock_data(stock, start_date, end_date) for stock in stocks}
data
```

```python

# Convert to DataFrame
df = pd.DataFrame(data)

# Reset the index to use for plotting
df.reset_index(inplace=True)
df
```

```python
# Plotting all on same plot
plt.figure(figsize=(14, 8))
for stock in stocks:
    sns.lineplot(x='Date', y=stock, data=df, label=stock)

# Customize the plot
plt.title('Stock Prices Over 5 Years')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend(title='Stock')
plt.grid(True)

# Show the plot
plt.show()
```
